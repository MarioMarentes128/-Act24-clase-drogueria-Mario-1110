## Clase Drogueria
Programación Orientada a Objetos (POO) que estamos viendo, la "Droguería" no es solo un local que vende medicinas, sino que funciona como un modelo o concepto abstracto.
![texto del vínculo](https://raw.githubusercontent.com/MarioMarentes128/-Act24-clase-drogueria-Mario-1110/refs/heads/main/infoclasedrogueria.png)


```python
class Drogueria:
    def __init__(self, nombre, inventario, encargado, domicilios):
        # Atributos de la clase
        self.nombre = nombre
        self.inventario = inventario
        self.farmaceuticoEncargado = encargado
        self.serviciosDomicilio = domicilios

    # Función 1: Surtir Receta
    def surtirReceta(self, medicamento):
        print(f"\n[Surtir] En la droguería {self.nombre} se ha surtido el medicamento: {medicamento}.")

    # Función 2: Asesorar Dosis
    def asesorarDosis(self, medicamento, dosis):
        print(f"[Asesoría] El farmacéutico {self.farmaceuticoEncargado} recomienda para {medicamento} una dosis de: {dosis}.")

    # Función 3: Controlar Caducidad
    def controlarCaducidad(self):
        print(f"[Control] Revisando el inventario de {self.inventario} artículos... ¡Todo está vigente!")

# --- PARTE INTERACTIVA ---

def crear_drogueria(numero):
    print(f"\n--- Ingrese los datos para la Droguería #{numero} ---")
    nom = input("Nombre de la droguería: ")
    inv = input("Cantidad de productos en inventario: ")
    enc = input("Nombre del farmacéutico encargado: ")
    dom = input("¿Cuenta con servicio a domicilio? (Si/No): ")

    # Retornamos el objeto creado con los datos del usuario
    return Drogueria(nom, inv, enc, dom)

# Creación de los 2 objetos solicitados
drogueria1 = crear_drogueria(1)
drogueria2 = crear_drogueria(2)

# --- EJECUCIÓN DE LAS FUNCIONES (Mensajes de ejemplo) ---

# Acciones para la primera droguería
drogueria1.surtirReceta("Paracetamol")
drogueria1.asesorarDosis("Paracetamol", "500mg cada 8 horas")
drogueria1.controlarCaducidad()

# Acciones para la segunda droguería
drogueria2.surtirReceta("Amoxicilina")
drogueria2.asesorarDosis("Amoxicilina", "1 tableta al día")
drogueria2.controlarCaducidad()

print("Programa Tterminado")
print("Mario Yañez NL = 53")
```
